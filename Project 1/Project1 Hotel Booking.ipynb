object data type means categorical column

now we need to see the categorical columns. So use describe function( use for numeric values) with include because we only wanted to see object attributes. 

```python
df. describe(include = 'object')
```

now I wanted to see that unique values in categorical columns

```python
unique_category = df. describe(include = 'object')

for col in unique_category.columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

Check null values

```python
df.isnull().sum()
```

now remove agent & company column , also remove null values data in country column or where the value is null

inplace=True --> use for changing data frame

```python
df.drop(['company', 'agent'], axis=1 , inplace= True)
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df = df[df['adr']< 5000]
```

```python
df.describe()
```

Data Analysis & Data Visalization

```python
canceled_perc = df['is_canceled'].value_counts(normalize = True)  #normalize returns percentage
print(canceled_perc)

plt.figure(figsize= (5,4))
plt.title('Reservation Status Count')
plt.bar(['Canceled', 'Not Canceled'],df['is_canceled'].value_counts())
plt.show()
```

now show how much both hotels have cancelation ratio

```python
plt.figure(figsize =(8,4))
ax1 = sns.countplot(x= 'hotel', hue= 'is_canceled', data = df, palette= 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in Differ Hotel', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['canceled', 'not canceled'])
plt.show()
```
